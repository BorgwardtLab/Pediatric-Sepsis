# choose the optimal K


```python
Mks = []
Aks = []
assignments = pd.DataFrame(columns=[2, 3, 4, 5, 6])
for k in [2, 3, 4, 5, 6]:
    physio_view = pd.read_csv("data/ThreeViewsKCC_physio_K_{}.csv".format(k), index_col=0)
    proteome_view = pd.read_csv("data/ThreeViewsKCC_proteome_K_{}.csv".format(k), index_col=0)
    contextual_view = pd.read_csv("data/ThreeViewsKCC_contextual_K_{}.csv".format(k), index_col=0)
    Xs = [physio_view, contextual_view, proteome_view]
    cons = ConsensusCluster(MultiviewKMeans, k, k+1, 100, resample_proportion=0.8)
    cons.fit(Xs)
    assert cons.Mk.shape[0] == 1
    Mks.append(cons.Mk[0])
```

```python
plt.figure(figsize=(12, 9))
colors = ['b', 'r', 'g', 'c', 'y']
for k in [2, 3, 4, 5, 6]:
    plt.hist(Mks[k-2].ravel(), bins=20, color=colors[k-2], alpha=0.3,
             weights=np.ones(len(Mks[k-2].ravel())) / len(Mks[k-2].ravel()),
             label=k, histtype='stepfilled')
plt.title('Consensus distribution', fontsize=25)
plt.legend()
plt.show()
```

```python
Aks = []
plt.figure(figsize=(12, 9))

for k in [2, 3, 4, 5, 6]:
    hist, bins = np.histogram(Mks[k-2].ravel(), bins=50)
    pdf = hist / sum(hist)
    cdf = np.cumsum(pdf)
    acdf = np.sum(h*(b-a) for b, a, h in zip(bins[1:], bins[:-1], cdf))
    Aks.append(acdf)
    plt.plot(bins[1:], cdf, label=k)
    
plt.legend(fontsize=16)
plt.title('CDF: PhysioContextualProteomeViews', fontsize=25)
plt.ylim(-0.05, 1.05)
plt.tight_layout()
plt.tight_layout()
plt.savefig('CDF plots/CDF PhysioContextualProteomeViews.png', dpi=300)
plt.show()
assignments.to_csv('data/PhysioContextualProteomeViewsAssignments.csv')
```

```python
delta_Aks = []
for k in range(len(Aks)):
    if k != 0:
        Aa = Aks[k-1]
        Ab = Aks[k]
        delta_Aks.append((Ab-Aa)/Aa)
    else:
        delta_Aks.append(Aks[k])

plt.figure(figsize=(9, 9))
plt.plot([2,3,4,5,6], delta_Aks, marker='*')
plt.xticks([2,3,4,5,6])
plt.title('Delta area', fontsize=25)
plt.show()
```

```python
for i in range(1, k + 1):
    for j in range(i + 1, k + 1):
        if j <= k:
            
            group1 = physio[physio['cluster'] == i]
            group2 = physio[physio['cluster'] == j]
            mean_value = pd.concat([group1, group2])[physio.columns[:-1]].mean()
            diff1 = group1[physio.columns[:-1]].mean() - mean_value
            diff2 = group2[physio.columns[:-1]].mean() - mean_value
            sorted_cols = diff1.sort_values().index

            plt.figure(figsize=(20,12))
            plt.plot(diff1[sorted_cols].values, c='b', label='cluster {}'.format(i))
            plt.plot(diff2[sorted_cols].values, c='r', label='cluster {}'.format(j))
            plt.axhline(y=0, color='black', linestyle='--')
            plt.xticks(np.arange(len(sorted_cols)), sorted_cols, rotation=90, fontsize=20)
            plt.yticks(rotation=90, fontsize=30)
            plt.title('Cluster {} vs {}'.format(i, j), fontsize=35)
            plt.tight_layout()
            plt.show()
```

## agreement between clusters


```python

proteome_cluster = pd.read_csv('Clustering/raw_km/ThreeViewsKCC_proteome_K_{}/ClusterAssignment.csv'.format(k), 
                          index_col=0)['cluster_{}'.format(k)]
physio_cluster = pd.read_csv('Clustering/raw_km/ThreeViewsKCC_physio_K_{}/ClusterAssignment.csv'.format(k), 
                               index_col=0)['cluster_{}'.format(k)]
contextual_cluster = pd.read_csv('Clustering/raw_km/ThreeViewsKCC_contextual_K_{}/ClusterAssignment.csv'.format(k), 
                               index_col=0)['cluster_{}'.format(k)]

agreements = pd.concat([proteome_cluster, physio_cluster, contextual_cluster], 1)
agreements.columns = ['Proteome', 'Physio', 'Context']
agreements['Multi'] = bcday['cluster']
```

```python
from matplotlib import colors

def background_gradient(s, m, M, cmap='PuBu', low=0, high=0):
    rng = M - m
    norm = colors.Normalize(m - (rng * low),
                            M + (rng * high))
    normed = norm(s.values)
    c = [colors.rgb2hex(x) for x in plt.cm.get_cmap(cmap)(normed)]
    return ['background-color: %s' % color for color in c]
```

```python
clusters = ['Proteome', 'Physio', 'Context', 'Multi']
assignments = [1, 2, 3]
combs = []
for cluster in clusters:
    for assignment in assignments:
        combs.append([cluster, assignment])
consensus_mat = pd.DataFrame()
for item_i in combs:
    for item_j in combs:
        index_1 = agreements[agreements[item_i[0]] == item_i[1]].index
        index_2 = agreements[agreements[item_j[0]] == item_j[1]].index
        len_overlap = len(list(set(index_1).intersection(set(index_2))))
        len_union = len(set(index_1).union(set(index_2)))
        consensus_mat.loc['{}_{}'.format(item_i[0], item_i[1]),
                          '{}_{}'.format(item_j[0], item_j[1])] = len_overlap / len_union
```

```python
consensus_mat = consensus_mat.astype(float)
consensus_mat_np = consensus_mat.values
np.fill_diagonal(consensus_mat_np, 0)
consensus_mat[consensus_mat.columns]= consensus_mat_np
consensus_mat.style.set_table_attributes('style="font-size: 12px"').set_precision(3).apply(background_gradient, cmap='Blues', 
               m=consensus_mat.min().min(),
               M=consensus_mat.max().max(),
               low=0,
               high=0.2)
```

```python
for assignment in ['Proteome', 'Physio', 'Context']:
    bcday['cluster'] = agreements[assignment]
    cons = [ 'cons05.resp', 'cons05.cvs', 'cons05.cns',
              'cons05.ren', 'cons05.hep', 'cons05.hem']
    overall_mean = bcday[cons +  ['cons05.score', 'ccc.summary', 'age.at.bc', 'sex', 'death.30.bc', 'picu' ]].mean()
    report = bcday.groupby('cluster').mean()[cons + \
                                       ['cons05.score', 'ccc.summary', 'age.at.bc', 'sex', 'death.30.bc', 'picu' ]]
    report['size'] = bcday.groupby('cluster').size().astype(str)
    report.loc['overall'] = overall_mean.tolist() + [str(len(bcday))]
    pd.set_option("display.precision", 3)
    display(report.T)
```
