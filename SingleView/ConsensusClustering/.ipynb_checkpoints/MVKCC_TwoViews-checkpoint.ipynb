# F-statistics

```python
def cal_F_stat(data, K):
    SSB = (
        ((data.groupby("cluster").mean() - data.drop("cluster", 1).mean()) ** 2).T
        * data.groupby("cluster").size().values
    ).sum(1)

    SSW = []
    for k in sorted(data["cluster"].unique()):
        cluster = data[data["cluster"] == k].drop("cluster", 1)
        diff = (cluster - data.groupby("cluster").mean().loc[k]) ** 2
        SSW.append(diff)
    SSW = pd.concat(SSW).sum()
    coef = (len(data) - K) / (K - 1)

    F_stat = (coef * (SSB / SSW)).sort_values(ascending=False)
    return F_stat
```

```python
data = copy.deepcopy(concated_view)
clusters = list(assignments["{} clusters".format(K)])
data["cluster"] = clusters
F_stat_true = cal_F_stat(data, K)
print(F_stat_true.sort_values(ascending=False).iloc[:10])
```

### bootstraping 

```python
all_F_stat = []
num_var = data.shape[1] - 1
n_test = 19 * num_var
for i in tqdm(range(n_test)):
    data_run = copy.deepcopy(data)
    rs = np.random.RandomState(i)
    data_run["cluster"] = rs.randint(1, K + 1, size=len(data_run))
    F_stat = cal_F_stat(data_run, K)
    all_F_stat.append(F_stat)
all_F_stat = pd.concat(all_F_stat, 1)
all_F_stat.columns = ["random {}".format(i + 1) for i in range(all_F_stat.shape[1])]

all_F_stat["true"] = F_stat_true
sign_F_stat = all_F_stat["true"].loc[
    all_F_stat["true"] > all_F_stat.quantile((1 - (0.05 / n_test)), axis=1)
]

print(sign_F_stat.sort_values(ascending=False))
```

```python
plt.figure(figsize=(16, 12))
weights = np.ones(all_F_stat.shape[0]) /all_F_stat.shape[0] 
for i in range(all_F_stat.shape[1] - 1):
    plt.hist(all_F_stat.iloc[:, i], bins=100, alpha=0.5, color='b', weights=weights,
             histtype='stepfilled')
plt.show()
```

```python
plt.figure(figsize=(16, 12))
weights = np.ones(all_F_stat.shape[0]) /all_F_stat.shape[0] 

plt.hist(all_F_stat.iloc[:, -1], bins=100, alpha=0.5, color='r', weights=weights,
             histtype='stepfilled', label='true')

plt.show()
```

## visualize clustering drivers

```python
target_var = sign_F_stat.sort_values(ascending=False).index[:10]
plt.figure(figsize=(30, 10))
for i, var in enumerate(target_var):
    plt.subplot(2, 5, i + 1)
    for k in range(1, K + 1):
        cluster = data[data['cluster'] == k]
        plt.hist(cluster[var], bins=25, alpha=0.5, density=True, 
                 histtype='stepfilled', label='cluster {}'.format(k))
    plt.legend()
    plt.title(var, fontsize=15)
plt.show()
```
